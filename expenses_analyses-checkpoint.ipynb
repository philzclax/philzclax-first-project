```python
import pandas as pd
```

```python
data = {
    "category": ["food", "snacks", "transport", "rent"],
    "amount": [2000, 1500, 10000, 500]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
df.groupby("category")["amount"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("category")["amount"].sum().plot(kind="bar")
plt.title("Expenses by Category")
plt.ylabel("Total Amount")
plt.xlabel("Category")
plt.show()
```

```python
plt.savefig("expenses_by_category.png")
```

```python
import matplotlib.pyplot as plt

# Group and sum data
data = df.groupby("category")["amount"].sum()

# Plot the data
data.plot(kind="bar")  

# Optional: add title and labels
plt.title("Expenses by Category")
plt.xlabel("Category")
plt.ylabel("Total Amount")

# Save figure
plt.savefig("expenses_by_category.png")

# Show figure
plt.show()
```
